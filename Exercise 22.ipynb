```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
df=pd.read_csv('/content/drive/MyDrive/Irwin data analysis portfolio/data/data/sat-scores.csv',
               usecols=['Year','State.Code','Total.Math','Total.Test-takers', 'Total.Verbal'])
df=df.set_index(['Year','State.Code'])

df.loc[2005, 'Total.Test-takers'].sum()# how many people took the SAT in 2005?

df.loc[(2010, ['NY', 'NJ', 'MA', 'IL']), 'Total.Math'].mean()# the average SAT math score in 2010 from New York, New Jersey, Massachussetts, Illinois

df.loc[([2012, 2013, 2014, 2015], ['AZ','CA','TX']), 'Total.Math'].mean().round(2)# the average SAT math score in 2012, 2013, 2014, 2015 in Arizona, California, Texas
```
