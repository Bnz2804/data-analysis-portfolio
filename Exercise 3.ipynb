```python
import pandas as pd
import numpy as np
g=np.random.default_rng(0)# we want the tens of each number in the series
s=pd.Series(g.integers(0,100,10))#the integers attribute is not applicable to a numpy array
Tens_digit=(s//10)#// is floor division it divides two numbers and rounds the result down to the nearest integer
print(Tens_digit)
```

```python
import pandas as pd
import numpy as np
g=np.random.default_rng(0)
s=pd.Series(g.integers(0,10000,10))#given a range this portion of the code generates random numbers constituting a series
Tens_digit=s.astype(str).str.get(-2).fillna("0").astype(int)# s.astype(str): converts the pandas series to a string to use string methods for extracting the second to last digit
print(Tens_digit)                                           #.str.get(-2): extracts the second to last character of each string in the series;-2 is used to access the second last character from the end
                                                            #.fillna("0"): fills any missing(NaN)values with the string "0"
                                                            #.astype(int): converts the resulting string to an integer
```

```python
import pandas as pd
import numpy as np
g=np.random.default_rng(0)# here we assigned to a variable g
#np.random.default_rng() is the recommended way to create a random number generator in NumPy. It gives us access to various methods for generating random numbers.
#(0) is the seed value. Using a seed ensures that the random numbers generated will be the same every time you run the code.
s=pd.Series(g.random(10)*1000)# Use g.random(number of elements) to generate random floats, then multiply by the desired range(here 1000).

s=s.loc[s.astype(int)%2==0] # Use astype(int) to convert elements to integers for comparison

print(f"{s.astype(int)}")#s.astype(int): This converts the elements of the Series s to integers, effectively extracting the integer component of each float.
    #% 2 == 0: This applies the modulo operator (%) to check if the integer component is divisible by 2. If the remainder is 0, it means the integer component is even.
    #s[...]: This uses boolean indexing to select only the elements of the original Series s where the condition (even integer component) is True.
```
