```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
df_2019_jul=pd.read_csv('/content/nyc_taxi_2020-07.csv',
                       usecols=['passenger_count', 'total_amount', 'payment_type'])
df_2019_jul['year']=2019
df_2020_jul=pd.read_csv('/content/nyc_taxi_2020-07.csv',
                       usecols=['passenger_count', 'total_amount', 'payment_type'])
df_2020_jul['year']=2020
df=pd.concat([df_2019_jul, df_2020_jul])

(df.loc[df['year']==2019,'total_amount'].count()-df.loc[df['year']==2020,'total_amount'].count())#how many rides were taken in 2019 versus 2020?

(df.loc[df['year']==2019,'total_amount'].sum().round(2)-df.loc[df['year']==2020,'total_amount'].sum().round(2))#how much less money did taxi drivers make as a result?

df.loc[(df['year']==2019) & (df['passenger_count']>1), 'passenger_count'].count() / df.loc[df['year']==2019,'payment_type'].count()#compare the proportion (not a raw number) of multiperson taxi rides in 2019 with those in 2020

df.loc[(df['year']==2020) & (df['passenger_count']>1), 'passenger_count'].count() / df.loc[df['year']==2020, 'payment_type'].count()# compare the proportion (not a raw number) of multiperson taxi rides in 2019 with those in 2020

print(df.loc[(df['year']==2019) & (df['payment_type']==2), 'payment_type'].count() / df.loc[df['year']==2019, 'payment_type'].count())# were people more or less likely to use cash during the pandemic, given that we were trying to avoid physical contact

print(df.loc[(df['year']==2020) & (df['payment_type']==2), 'payment_type'].count() / df.loc[df['year']==2020, 'payment_type'].count())
```
