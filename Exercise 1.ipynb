```python
import pandas as pd
import numpy as np
import math
g=np.random.default_rng(0)# numpy's random array generator
months="Sep Oct Nov Dec Jan Fev Mar Apr May Jun".split()
s=pd.Series(g.integers(70,101,10), index=months)#the random method takes three elements: the minimum value, the maximum value+1 and the length of the array
print (s)
print(f'Yearly average:{s.mean()}')#What is the yearly average of the recorded scores for the academic year?
First_half_average=s.loc["Sep":"Jan"].mean()#What is the first half average? P.S: Use "" instead of '' in print(f{s.loc[]})
Second_half_average=s.loc["Fev":"Jun"].mean()#What is the second half average?
Standard_dev=math.sqrt(((s-s.mean())**2).sum()/s.count())
Student_five_highest_scores=s.nlargest(5)
Round_scores=(s/10).round().astype(int)*10# we divide each element of the series by 10 first, second we round it up,third we make the outcome an integer, finally we multiply it by 10
print(f'First half average:{First_half_average}')
print(f'Second half average:{Second_half_average}')
print(f'Improvement:{Second_half_average-First_half_average:.4f}')#What is the improvement by half year P.S: to round our outcome to any nÂ° digits after the decimal point, use :(.2f)
print(f'Student highest score is {s.max()} achieved in {s.idxmax()}')
print("The student's five highest scores are:\n" +str(Student_five_highest_scores))
print("This student's scores rounded to the nearest 10 are :\n"+str(Round_scores))
print(f'Standard deviation:{Standard_dev:.4f}')
```
