```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame

#Passing a dict of lists. Each key represents one coolumn, and the values(lists) are each column's values
df = DataFrame (
    {'product_id':[23, 96, 97, 15, 87],
     'name':['computer','Python Workout','Pandas Workout','banana','sandwich'],
     'wholesale_price':[500, 35, 35, 0.5, 3.0],
     'retail_price':[1000, 75, 75, 1, 5],
     'sales':[100, 1000, 500, 200, 300]
     },index = list('01234'))
#list of dicts
new_products= DataFrame(
    [{'product_id':24, 'name':'phone', 'wholesale_price':200, 'retail_price':500},
     {'product_id':16, 'name':'apple', 'wholesale_price':0.5, 'retail_price':1},
   {'product_id':17, 'name':'pear', 'wholesale_price':0.6, 'retail_price':1.2} ],index=range(5,8))

df=pd.concat([df,new_products])
df.loc[[5,6,7], 'sales']=[100, 200, 75]

df.loc[8,['product_id','name', 'wholesale_price','retail_price','sales' ]]=[44, 'wallet', 10, 20, 50]
df['department']=['electronics', 'books', 'books', 'food', 'food', 'electronics', 'food', 'food', 'others']

df['total_sales']= (df['retail_price']-df['wholesale_price'])*df['sales']

df['descriptive_stats']= ''
food_stats= df[df['department']=='food']['total_sales'].describe()# in the department column where the product is 'food' give me descriptive statistics in the total sales column

food_stats_str=food_stats.to_string()#all descriptive statistics come initially as floats, this line converts all of them to strings to integrate them to the descriptive stats column created initially
df.loc[df['department']=='food', 'descriptive_stats']=food_stats_str

df['current_net']=df['total_sales'].cumsum()#the cumulative sum of each total sales
df.loc['Total']=df[['wholesale_price', 'retail_price', 'sales', 'total_sales']].sum()
df.loc['Total']=df.loc['Total'].fillna('')
df
```
