```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
temp_data=pd.read_csv('/content/nyc-temps.txt',header=None).squeeze()
hour_data=[0, 3, 6, 9, 12, 15, 18, 21]*(len(temp_data)//8)
hour_data.extend(hour_data[:len(temp_data)%8])
#if len(temp_data)%8!=0:

df=DataFrame({'temp':temp_data,'hour':hour_data})
df.loc[df['hour'].isin([3,6]),'temp']=np.NaN#NaN is not a built-in keyword in Python. It is a special floating-point value representing "Not a Number" and is typically accessed through the numpy library as np.nan
df=df.interpolate()
df.describe()
#df['temp'].describe()
```
