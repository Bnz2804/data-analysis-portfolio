```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/taxi-distance.csv',header=None).squeeze()

pd.cut(df,
       bins=[0, 2, 10, df.max()],
       include_lowest=True,
       labels=['short','medium','long']).value_counts(normalize=True)
mean=df.mean()
median=df.median()
print(f'The mean trip distance is {mean} and the median trip distance is {median}')
print('The mean is greater than the median, the data is skewed to the left')
#df=pd.read_csv('/content/taxi-distance.csv',header=None
#categories=df.astype(str)
#categories.loc[:]='medium'
#categories.loc[df<=2]='short'
#categories.loc[df>10]='long'
#categories.value_counts()
```
