```python
import pandas as pd
import numpy as np
from pandas import Series
days='Sun Mon Tue Wed Thu Fri Sat'.split()#.split() is a string method that splits the string into a list of substrings based on a delimiter. By default, it splits on spaces, and doesn't work on lists
g=np.random.default_rng(0)
s=Series(g.normal(20,5,28), index=days*4).round().astype(np.int8)#.round(0): Instead of int, we use 0, which indicates that we want to round the values to the nearest integer.
s.loc['Mon'].mean()#.astype(np.int8): This converts the data type of the numbers in the Series to an 8-bit integer
```
