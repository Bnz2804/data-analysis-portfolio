```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
df=pd.read_csv('/content/nyc_taxi_2019-01.csv',
              usecols=['passenger_count', 'trip_distance', 'payment_type', 'total_amount']).squeeze()
df.loc[df['passenger_count']>8, 'passenger_count'].count()# How many taxi rides had more than eight passengers?

df.loc[df['passenger_count']==0,'passenger_count'].count()# how many taxi rides in January 2019 had zero passengers?

df.loc[(df['payment_type']==2) & (df['total_amount']>1000), 'total_amount'].count()#How many rides that month were paid for in cash and had a total_amount of more than $1,000?

df.loc[df['total_amount']<0,'total_amount'].count()#how many rides cost less than $0

df.loc[(df['trip_distance']<df['trip_distance'].mean()) & (df['total_amount']>df['total_amount'].mean()),'trip_distance'].count()#how many trips traveled a below-average distance but cost an above-average amount?

df.loc[df['payment_type']==1,'payment_type'].count()#how many rides were paid by card

cash_percentage=(df['payment_type']==2).sum()/(df['payment_type'].sum())

card_percentage=(df['payment_type']==1).sum()/(df['payment_type'].sum())

print(f"The percentage of riders paying cash is:{cash_percentage}, while the percentage of riders using cards is:{card_percentage}")
```
