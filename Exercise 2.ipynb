```python
import pandas as pd
import numpy as np
import math
g=np.random.default_rng(0)
month="Sep Oct Nov Dec Jan Feb Mar Apr May Jun".split()
s=pd.Series(g.integers(40,60,10), index=month)
Standard_deviation= math.sqrt(((s-s.mean())**2).sum()/s.count())
Addition=s+(80-s.mean())
Mean=s.mean(numeric_only=True)
Median=s.median(numeric_only=True)#for the s.mean and s.median method it is primordial to specify numeric_only=True to get a single value for the whole series
print("The new scores are: \n"+str(Addition))
Diff=Mean-Median
for index, score in s.items():#the for statement is much more indicated to compare each value in the series to a single value

  if score in range(int(s.mean()-Standard_deviation)):#if statements only work on a single value relative to another not on a series
    print(f"The student got a C, for {index} ")
  elif score in range(int (s.mean()+Standard_deviation)):#range only works on integers
    print(f"The student got a B, for {index} ")
  elif score>(s.mean()+Standard_deviation):
    print(f"The student got an A, for {index} ")
  elif score<(s.mean()-Standard_deviation):
    print(f"The student got a D, for {index} ")
  elif score in range(int(s.mean()-(2*Standard_deviation)), int(s.mean()+(2*Standard_deviation))):
    print(f"Months {index} saw above average scores")
print("The mean is "+str(Mean)+" and the median is "+str(Median)+" the difference is "+str(Diff))
if Diff ==0 :
  print("The distribution is normal ")
else:
  print("The data is skewed")
```
