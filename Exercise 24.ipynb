```python
import pandas as pd
import numpy as np

from pandas import Series, DataFrame
df=pd.read_csv('/content/drive/MyDrive/Irwin data analysis portfolio/data/data/olympic_athlete_events.csv',
               usecols=['Age','Height','Team','Year','Season','Sport', 'Medal'])

df=df.loc[df['Year']>=1980]
df=df.loc[df['Team'].isin(['Great Britain','France','United-States','Switzerland','China','India'])]

df.pivot_table(index='Year', columns='Team', values='Age').round()# compare the average age of players for each team, for all sports and all years

df.pivot_table(index='Sport', columns='Year', values='Height', aggfunc='max')# we want to find the tallest players in each sport from each year

df.dropna(subset='Medal').pivot_table(index='Year', columns='Team', values='Medal',aggfunc='size')# we want to determine how many medals each country received at each game PS::we could've simply left the dropna method and used the count() function and still gotten the same result

df.pivot_table(index=['Year','Season'], columns='Team',values='Medal',aggfunc='count')# how many medals each team won per year with the index including the year and the season

df.pivot_table(index='Year',columns='Team',values=['Age','Height'])# average age and average height per year per team

df.pivot_table(index=['Year','Season'], columns='Team', values=['Age','Height'])# the average age and average height per year , per team, broken up by year and season
```
