```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
g=np.random.default_rng(0)
s=Series(g.normal(0,100,100_000))
print(s.describe())
s_plus_aver=s.loc[s>s.mean()]
print(f'The mean of the values greater than the mean is {s_plus_aver.mean()}')
s_minus_aver=s.loc[s<s.mean()]
print(f'The mean of the values less than the mean is {s_minus_aver.mean()}')
s_plus_minus_aver=((s_plus_aver)+(s_minus_aver)).fillna(0).mean()
if s_plus_minus_aver != s.mean():
  print("The mean of the values greater or less than the standard mean is "+ str(s_plus_minus_aver)+". It is not equal to the standard mean of the values."+str(s.mean()))
else:
  print('The mean of the values greater or less than the standard mean is equal to the standard mean of the values.')
```
